```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('85cbb44bd7a7544ae9c2a71206d7ce59.xlsx', skiprows=5)
df.head() 
```

```python
df_akhir = df[-14:-3]
df_akhir.tail()
```

```python
#
df_akhir['NO.'] = df_akhir['Unnamed: 1']
df_akhir['NIM'] = df_akhir['Unnamed: 3']
df_akhir['NAMA'] = df_akhir['Unnamed: 5']
df_akhir['FAKULTAS'] = df_akhir['Unnamed: 7']
df_akhir['LOKASI'] = df_akhir['Unnamed: 10']
df_akhir['KECAMATAN'] = df_akhir['Unnamed: 12']
df_akhir['KABUPATEN'] = df_akhir['Unnamed: 15']
df_akhir.tail()
```

```python
#removing unnamed header
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.dropna(inplace=True)
df.tail()
```

```python
df_akhir = df_akhir.loc[:, ~df_akhir.columns.str.contains('^Unnamed')]
df_akhir.tail()
```

```python
#join both df
df = pd.concat([df, df_akhir])
df.tail()
```

```python
#finding duplicates
df[df.duplicated()]
```

```python
print(len(df))
df.drop('NO.', axis=1, inplace=True)
df_clean = df.drop_duplicates()
print(len(df_clean))
```

```python
#grouping by lokasi
df_grouped = df_clean.groupby('LOKASI')
df_grouped.head()
```

```python
#distinct per lokasi
df_grouped['LOKASI'].unique().count()
```

```python
#rata-rata per lokasi
df_grouped.agg('count')['NIM']
desa_summary = df_grouped.agg('count')['NIM'].sum()
total_desa = df_grouped['LOKASI'].unique().count()
rata_rata = desa_summary/total_desa
print(rata_rata)

minimal = df_grouped.agg('count')['NIM'].min()
print(minimal)

maksimal = df_grouped.agg('count')['NIM'].max()
print(maksimal)

#find minimal
print(df_grouped.agg('count')['NIM'].idxmin())

#find maximal
print(df_grouped.agg('count')['NIM'].idxmax())

#total nim per lokasi
print(df_grouped['NIM'].count().sort_values(ascending=True))

#cek kutampi
# print(df_grouped.get_group('KUTAMPI').sort_values(by='NIM'))
```
